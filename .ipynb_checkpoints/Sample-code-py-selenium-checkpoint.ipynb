```python
# Matplotを使った図形描画
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def run_plotting():

    # サンプルデータを作成する (仮のデータ)
    data = np.array([
        [101, 5, 13, 7],
        [102, 2, 14, 8],
        [103, 1, 15, 6],
        [104, 6, 7, 2],
        [105, 8, 6, 3],
        [106, 9, 2, 5]
    ])
    
    def plot_line_chart(ax, df):
        ax.plot(df['ID'], df['Value1'], marker='o', label='Value1')
        ax.plot(df['ID'], df['Value2'], marker='s', label='Value2')
        ax.plot(df['ID'], df['Value3'], marker='^', label='Value3')
        ax.set_title('Line Plot')
        ax.set_xlabel('ID')
        ax.set_ylabel('Values')
        ax.legend()
        ax.grid(True)

    def plot_bar_chart(ax, df):
        bar_width = 0.25
        indices = np.arange(len(df))
        ax.bar(indices, df['Value1'], width=bar_width, label='Value1')
        ax.bar(indices + bar_width, df['Value2'], width=bar_width, label='Value2')
        ax.bar(indices + 2 * bar_width, df['Value3'], width=bar_width, label='Value3')
        ax.set_title('Bar Plot')
        ax.set_xlabel('ID')
        ax.set_ylabel('Values')
        ax.set_xticks(indices + bar_width)
        ax.set_xticklabels(df['ID'])
        ax.legend()
        ax.grid(True)

    def plot_scatter_chart(ax, df):
        ax.scatter(df['ID'], df['Value1'], marker='o', label='Value1')
        ax.scatter(df['ID'], df['Value2'], marker='s', label='Value2')
        ax.scatter(df['ID'], df['Value3'], marker='^', label='Value3')
        ax.set_title('Scatter Plot')
        ax.set_xlabel('ID')
        ax.set_ylabel('Values')
        ax.legend()
        ax.grid(True)

    def plot_box_chart(ax, df):
        ax.boxplot([df['Value1'], df['Value2'], df['Value3']], labels=['Value1', 'Value2', 'Value3'])
        ax.set_title('Box Plot')
        ax.set_xlabel('Values')
        ax.set_ylabel('Distribution')
        ax.grid(True)

    def plot_histogram(ax, df):
        ax.hist(df['Value1'], bins=5, alpha=0.7, label='Value1')
        ax.hist(df['Value2'], bins=5, alpha=0.7, label='Value2')
        ax.hist(df['Value3'], bins=5, alpha=0.7, label='Value3')
        ax.set_title('Histogram')
        ax.set_xlabel('Values')
        ax.set_ylabel('Frequency')
        ax.legend()

    # DataFrameを作成
    df = pd.DataFrame(data, columns=['ID', 'Value1', 'Value2', 'Value3'])

    # プロットの配置を作成
    fig, axs = plt.subplots(2, 3, figsize=(16, 9))

    # 各種プロットを描画
    plot_line_chart(axs[0, 0], df)
    plot_bar_chart(axs[0, 1], df)
    plot_scatter_chart(axs[0, 2], df)
    plot_box_chart(axs[1, 0], df)
    plot_histogram(axs[1, 1], df)

    # レイアウトを調整して表示
    plt.tight_layout()
    plt.show()
    
#     # テスト
# run_plotting()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def run_standardizing():

    # サンプルデータを作成
    data = np.array([
        [101, 5, 13, 7],
        [102, 2, 14, 8],
        [103, 1, 15, 6],
        [104, 6, 7, 2],
        [105, 8, 6, 3],
        [106, 9, 2, 5]
    ])

    # DataFrameを作成
    df = pd.DataFrame(data, columns=['ID', 'Value1', 'Value2', 'Value3'])

    # 標準正規分布に変換
    standardized_df = df.copy()
    for col in ['Value1', 'Value2', 'Value3']:
        mean = standardized_df[col].mean()
        std = standardized_df[col].std()
        standardized_df[col] = (standardized_df[col] - mean) / std

    # データを整形して箱ひげ図を作成
    melted_df = df.melt(id_vars=['ID'], value_vars=['Value1', 'Value2', 'Value3'], var_name='Variable', value_name='Value')
    melted_standardized_df = standardized_df.melt(id_vars=['ID'], value_vars=['Value1', 'Value2', 'Value3'], var_name='Variable', value_name='Value')

    # 2つの箱ひげ図を縦に並べて表示（X軸とY軸を入れ替え）
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # figsizeを調整して適切なサイズにする

    # 平均値の表示方法を設定
    meanprops = {"marker": "o", "markerfacecolor": "black", "markeredgecolor": "black"}

    # 元のデータの箱ひげ図（縦横回転）
    sns.boxplot(y='Variable', x='Value', data=melted_df, ax=axs[0], orient='h', showmeans=True, meanprops=meanprops, palette='Set2')
    axs[0].set_title('Original Data')

    # 標準化後のデータの箱ひげ図（縦横回転）
    sns.boxplot(y='Variable', x='Value', data=melted_standardized_df, ax=axs[1], orient='h', showmeans=True, meanprops=meanprops, palette='Set2')
    axs[1].set_title('Standardized Data')

    plt.tight_layout()
    plt.show()

# Test
# run_standardizing()
```

```python
# Tf-Idf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfTransformer
from scipy.stats import norm

def run_tfidf():

    # サンプルデータを作成
    data = np.array([
        [101, 5, 13, 7],
        [102, 2, 14, 8],
        [103, 1, 15, 6],
        [104, 6, 7, 2],
        [105, 8, 6, 3],
        [106, 9, 2, 5]
    ])

    # DataFrameを作成
    df = pd.DataFrame(data, columns=['ID', 'Value1', 'Value2', 'Value3'])

    # TF-IDFの計算
    tfidf_transformer = TfidfTransformer()
    tfidf_data = tfidf_transformer.fit_transform(df[['Value1', 'Value2', 'Value3']]).toarray()
    tfidf_df = pd.DataFrame(tfidf_data, columns=['Value1', 'Value2', 'Value3'])

    # TF-IDF値を標準正規分布に正規化
    tfidf_df_norm = pd.DataFrame()
    for col in ['Value1', 'Value2', 'Value3']:
        # 平均0、標準偏差1の標準正規分布に変換
        tfidf_df_norm[col] = norm.cdf(tfidf_df[col])

    # 横に並べて表示
    fig, axs = plt.subplots(1, 2, figsize=(18, 6))

    # 元のTF-IDFのヒートマップ
    sns.heatmap(tfidf_df, cmap='Reds', annot=True, fmt=".2f", cbar_kws={'label': 'TF-IDF Value'}, ax=axs[0])
    axs[0].set_title('TF-IDF Heatmap')

    # 標準正規分布に正規化したTF-IDFのヒートマップ
    sns.heatmap(tfidf_df_norm, cmap='Reds', annot=True, fmt=".2f", cbar_kws={'label': 'Standard Normal TF-IDF Value'}, ax=axs[1])
    axs[1].set_title('Standard Normal TF-IDF Heatmap')

    plt.tight_layout()
    plt.show()

# Test
# run_tfidf()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity

def run_cosine_similarity_comparison():

    # サンプルデータを作成
    data = np.array([
        [101, 5, 13, 7],
        [102, 2, 14, 8],
        [103, 1, 15, 6],
        [104, 6, 7, 2],
        [105, 8, 6, 3],
        [106, 9, 2, 5]
    ])

    # DataFrameを作成
    df = pd.DataFrame(data, columns=['ID', 'Value1', 'Value2', 'Value3'])

    # コサイン類似度の計算 (正規化前)
    cosine_sim_original = cosine_similarity(df[['Value1', 'Value2', 'Value3']])

    # データの正規化
    scaler = StandardScaler()
    df_normalized = df.copy()
    df_normalized[['Value1', 'Value2', 'Value3']] = scaler.fit_transform(df[['Value1', 'Value2', 'Value3']])

    # コサイン類似度の計算 (正規化後)
    cosine_sim_normalized = cosine_similarity(df_normalized[['Value1', 'Value2', 'Value3']])

    # TF-IDFの計算 (正規化前)
    transformer_original = TfidfTransformer()
    tfidf_matrix_original = transformer_original.fit_transform(df[['Value1', 'Value2', 'Value3']])
    cosine_sim_tfidf_original = cosine_similarity(tfidf_matrix_original)

    # TF-IDFの計算 (正規化後)
    transformer_normalized = TfidfTransformer()
    tfidf_matrix_normalized = transformer_normalized.fit_transform(df_normalized[['Value1', 'Value2', 'Value3']])
    cosine_sim_tfidf_normalized = cosine_similarity(tfidf_matrix_normalized)

    # ヒートマップの描画
    fig, axs = plt.subplots(2, 2, figsize=(14, 14))

    # 正規化前のコサイン類似度ヒートマップ
    sns.heatmap(cosine_sim_original, cmap='YlGnBu', annot=True, fmt=".2f", cbar_kws={'label': 'Cosine Similarity'}, ax=axs[0, 0], xticklabels=df['ID'], yticklabels=df['ID'])
    axs[0, 0].set_title('Original Cosine Similarity Heatmap')

    # 正規化後のコサイン類似度ヒートマップ
    sns.heatmap(cosine_sim_normalized, cmap='YlGnBu', annot=True, fmt=".2f", cbar_kws={'label': 'Cosine Similarity'}, ax=axs[0, 1], xticklabels=df['ID'], yticklabels=df['ID'])
    axs[0, 1].set_title('Normalized Cosine Similarity Heatmap')

    # 正規化前のTF-IDFを使用したコサイン類似度ヒートマップ
    sns.heatmap(cosine_sim_tfidf_original, cmap='YlGnBu', annot=True, fmt=".2f", cbar_kws={'label': 'Cosine Similarity'}, ax=axs[1, 0], xticklabels=df['ID'], yticklabels=df['ID'])
    axs[1, 0].set_title('Cosine Similarity using TF-IDF (Original)')

    # 正規化後のTF-IDFを使用したコサイン類似度ヒートマップ
    sns.heatmap(cosine_sim_tfidf_normalized, cmap='YlGnBu', annot=True, fmt=".2f", cbar_kws={'label': 'Cosine Similarity'}, ax=axs[1, 1], xticklabels=df['ID'], yticklabels=df['ID'])
    axs[1, 1].set_title('Cosine Similarity using TF-IDF (Normalized)')

    plt.tight_layout()
    plt.show()

# Test
# run_cosine_similarity_comparison()
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

def run_clustering(nclusters, IsPlot, initial_point_id=None):

    # サンプルデータを作成する (仮のデータ)
    data = np.array([
        [101, 5, 13, 7],
        [102, 2, 14, 8],
        [103, 1, 15, 6],
        [104, 6, 7, 2],
        [105, 8, 6, 3],
        [106, 9, 2, 5]
    ])

    # 社員番号を除外したデータを作成する
    data_without_id = data[:, 1:]

    if initial_point_id is not None:
        print(initial_point_id)
        # 指定されたIDに対応するデータポイントを取得する
        initial_point = data[data[:, 0] == initial_point_id][:, 1:]
        if initial_point.size == 0:
            raise ValueError(f"指定されたID {initial_point_id} に対応するデータポイントが見つかりません。")

        # 他の初期中心点をランダムに選択
        other_center = data_without_id[np.random.choice(data_without_id.shape[0], nclusters - 1, replace=False)]

        # 初期センターを設定
        initial_centers = np.vstack((initial_point, other_center))

        # k-meansクラスタリングを初期センターを指定して実行する
        kmeans = KMeans(n_clusters=nclusters, init=initial_centers, n_init=1)
    else:
        # k-meansクラスタリングをデフォルト設定で実行する
        kmeans = KMeans(n_clusters=nclusters)

    kmeans.fit(data_without_id)

    # 各データポイントがどのクラスタに属するかを取得する
    labels = kmeans.labels_

    # クラスタの中心点を取得する
    centers = kmeans.cluster_centers_

    # 各データポイントからクラスタの中心点までの距離を計算する
    distances = np.linalg.norm(data_without_id - centers[labels], axis=1)

    # 各データポイントの所属するクラスタとクラスタの中心点からの距離を表示する
    for i, (label, distance) in enumerate(zip(labels, distances)):
        print("社員番号 {} のデータはクラスタ {} に属しています。クラスタの中心点からの距離: {:.2f}".format(data[i, 0], label, distance))

    # 散布図を表示する
    if IsPlot:
        plt.scatter(data_without_id[:, 0], data_without_id[:, 1], c=labels, cmap='viridis')
        plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
        for i, txt in enumerate(data[:, 0]):
            plt.annotate(txt, (data_without_id[i, 0], data_without_id[i, 1]))
        plt.title('Cluster Analysis')
        plt.show()

# テスト
# run_clustering(nclusters=2, IsPlot=True)
# run_clustering(nclusters=3, IsPlot=True, initial_point_id=103)
```

```python
# Seleniumを使ったブラウザ操作
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

def run_browser(IsAutoInstall, IsHeadless):
    
    # Chrome起動オプションを設定（headless /通常）
    options = Options()
    if IsHeadless:
        print("Headlessモードでブラウザ起動します")
        options.add_argument('--headless')
        options.add_argument('--disable-gpu')
    else:
        print("通常モードでブラウザ起動します")

    # Chromeブラウザを起動
    if IsAutoInstall:
        # 自動インストール
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service, options=options)
    else:
        # 手動ダウンロードしたChromeドライバーのパスを指定
        if IsHeadless:
            chrome_driver_path = "./chromedriver/chromedriver_headless"
        else:
            chrome_driver_path = "./chromedriver/chromedriver"
        driver = webdriver.Chrome(executable_path=chrome_driver_path, options=options)

    # 指定したURLにアクセス
    driver.get("https://www.google.com")

    # 必要な処理を実行
    # ここに任意の処理を追加

    # ブラウザを終了
    driver.quit()
    if IsHeadless:
        print("Headlessモードのブラウザを終了しました")
    else:
        print("通常モードのブラウザを終了しました")
```

```python
# faster-whisperを使った文字起こし
from faster_whisper import WhisperModel

def run_trancribe(input_filepath, output_filepath):

    model_sizes = [
        "small",
#         "medium", 
#         "large-v1", 
#         "large-v2", 
#         "large-v3", 
#         "distil-large-v2", 
#         "distil-large-v3"
    ]
    model_size = model_sizes[0]

    # run on CPU with INT8
    model = WhisperModel(model_size, device="cpu", compute_type="int8")

    segments, info = model.transcribe(input_filepath, beam_size=5)

    print("Detected language '%s'" % (info.language))
    print("probability %f" % (info.language_probability))

    # 結果をバッファに蓄積
    buffer = []
    for segment in segments:
        buffer.append(f"[{segment.start:.2f}s -> {segment.end:.2f}s] {segment.text}\n")
    
    # バッファの内容をファイルに書き出し
    with open(output_filepath, "w", encoding="utf-8") as file:
        file.writelines(buffer)
```
